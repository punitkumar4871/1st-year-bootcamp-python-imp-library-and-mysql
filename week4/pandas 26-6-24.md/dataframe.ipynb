```python
#coming to the operations in pandas 
print("1. max()\n2. min()\n3. describe()\n4. mean()\n5. std()")
```

```python
#remember one thing while using mean there can be some mistake bcz of null value in the excel so to handnle this we use fillna method to handle null values in tpandas
#max()
print("the pokeon having tha max attack is",df["Attack"].max())
print("and the pokemon name is",df["Name"][df["Attack"]==190])
```

```python
#min()
print("the pokeon having tha min attack is",df["Attack"].min())
print("and the pokemon name is",df["Name"][df["Attack"]==5])
```

```python
#mean()
print("the average stats total point of pokemon till generation 6 is",df["Total"].mean())
```

```python
#std() used for standard deviation
print("the standard deviation  stats total point of pokemon till generation 6 is",df["Total"].std())
```

```python
#decribe it is used for printing all the statisic of all data which is in form of numbers
df.describe()
```

```python
#more conditional based selection using <,>,==
df[df["Defense"]>200] #this to ge tthe name of pokemon with defec=nce greater than 20
```
