# handling special values in pandas
#  using replace funcn


```python
import pandas as pd
import numpy as np
df=pd.read_excel("D:\sql sem 2\pandashandlingspecial.xlsx","Sheet1")
df
```

```python
'''how i can replace -99999(special) value to null value'''
new_df=df.replace(-99999,np.NaN)
new_df
#this can be good if only one special value is there
```

# for handling multiple special
# by list not recommened bcz if event has 0 and i make it null but 0 can be in wind also and its valid so use dictionary for working on specific columns

```python
new_df=df.replace([-99999,-88888,0],np.NaN)
new_df
#by list not recommened bcz if event has 0 and i make it null but 0 can be in wind also and its valid so use dictionary for working on specific columns
```

# using dictionary

```python
new_df=df.replace({
    "temperature":-99999,
    "windspeed":-88888,
    "event":0},np.NaN)

new_df
```

```python
'''question make nan ki jagah sunny in event and -99999 ki jagah nan
'''
new_df=df.replace({
     -99999:np.NaN,
     -88888:np.NaN,
      0:'sunny'
})
new_df
```

# coming to the principle of regx
basicallt regex do changement according to the pattern

```python
df1=pd.read_excel("D:\sql sem 2\pandas6regx.xlsx","Sheet1")
df1
```

```python
new_df=df.replace("[A-Za-z]",'',regex=True)
new_df
```

```python
#by this all character are gone consisting of event also
```

```python
# so to overcome this we use discovery
new_df=df.replace({
    'temperature':"[A-Za-z]",
    'windspeed':"[A-Za-z]"},''
    ,regex=True)
new_df
```
