# concat dataframe

```python

import pandas as pd
india_weather={
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,70]
}
india_weather=pd.DataFrame(india_weather)
india_weather
```

```python
us_weather={
    "city":["new york","chicago","oriando"],
    "temperature":[21,14,35],
    "humidity":[6,65,75]
}
us_weather=pd.DataFrame(us_weather)
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
#can take more than 2 argument
```

```python
#its using index as previous only let say we want continous index
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
'''if we dont want index instead of that we want someone more good is
'''
df=pd.concat([india_weather,us_weather],keys=["india","us"])
df
#it also give additional index
```

```python
df.loc["india"]
```

# now what if i want to add vertically not horizontally
# by axis


```python
temperature_df={
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30]
    
}
temperature_df=pd.DataFrame(temperature_df)
temperature_df
```

```python
windspeed_df={
    "city":["mumbai","delhi","bangalore"],
    "windspeed":[7,12,9]
    
}
windspeed_df=pd.DataFrame(windspeed_df)
windspeed_df
```

```python
g1=pd.concat([temperature_df,windspeed_df])
g1
```

```python
#what i want is that windspeed to act as a column
'''by axis'''

g1=pd.concat([temperature_df,windspeed_df],axis=1)
g1
#here if u use ignore index it will remove upper headersbcz axis is changed
```
